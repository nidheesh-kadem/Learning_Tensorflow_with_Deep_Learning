**What are activation functions?**

Activations functions like ReLU, Sigmoid, and Tanh introduce non-linearity, enabling the network to learn and approximate complex, non-linear decision boundaries in the data. This non-linearity allows the network to map complex inputs to specific outputs

Sigmoid:
Shape: S-shaped curve.
Range: Output is between 0 and 1.
Use: Commonly used in binary classification problems (e.g., to give probabilities).
Issue: It can cause the vanishing gradient problem, meaning gradients become very small, which slows down learning.
Example: Imagine you're predicting if an email is spam (0) or not spam (1). Sigmoid is great for this because its output is always between 0 and 1.

Tanh:
Shape: S-shaped curve (similar to Sigmoid but steeper).
Range: Output is between -1 and 1.
Use: Often used when we want the output to range from negative to positive, helping center the data.
Issue: Like Sigmoid, it also suffers from the vanishing gradient problem.
Example: Tanh is useful when you want to predict values that can be negative or positive, like temperature changes (-1 to 1).

ReLU (Rectified Linear Unit):
Shape: A straight line for positive values and flat (zero) for negative values.
Range: Output is between 0 and infinity (only positive values are passed; negatives are set to 0).
Use: Popular in deep learning because it helps avoid the vanishing gradient problem and speeds up learning.
Issue: It can cause the "dying ReLU" problem, where neurons stop learning if they only output zeros.
Example: ReLU is typically used in hidden layers of neural networks because it allows the model to learn faster and handle complex patterns.
